```python
# import packages 
import pandas as pd
import os 
import matplotlib.pyplot as plt
```

```python
# get files 
files = [f for f in os.listdir('../data/mimic-iv-clinical-database-demo-2.2/hosp/')]

# import data 
data = {}
table_names = []
for f in files:
    table = pd.read_csv(f"../data/mimic-iv-clinical-database-demo-2.2/hosp/{f}", compression='gzip', low_memory=False)
    table_name = f.split('.')[0]
    table_names.append(table_name)
    data[table_name] = table
```

# Table Identification

- omr: online medical record, outpatient info such as bp, wight, height, bmi
- admissions: info on the patient's admission to the hospital + demographics
    - links to patients on subject_id 
- d_icd_diagnoses: diagnosis codes, typically assigned at end of patient's stay
    - links to diagnoses_icd on icd_code & icd_version
- d_labitems: definitions for itemid associated with lab measurements
    - links to labevents on itemid
- diagnoses_icd: records all diagnoses a patient was billed for
- emar: electronic medicine administration record, medications given (implemented 2011-2013) 
    - links to emar_detail on emar_id
    - pharmacy on pharmacy_id
    - prescriptions on pharmacy_id
    - poe on poe_id
- emar_detail: info for medicine given to patient i.e., dose given
- labevents: all laboratory measurements for a patient
    - links to d_labitems on itemid
- **microbiologyevents**: records microbiology tests i.e., cultures + sensitivity results
    - links to admissions using subject_id, adittime, dischtime (captures labs proxmial to hospital stays)
- patients: age, geneder, date of death
- pharmacy: info on filled medications & courses
    - links to poe on poe_id
    - prescriptions on pharmacy_id
    - emar on pharmacy_id 

```python
mb = data['microbiologyevents']
plt.figure(figsize=(12, 6))
plt.hist(mb.spec_type_desc)
plt.xticks(rotation=70)
plt.show()
```
